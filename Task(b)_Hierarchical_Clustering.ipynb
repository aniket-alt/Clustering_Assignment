```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_moons, make_blobs
from sklearn.metrics import silhouette_score

# 1. GENERATE DATA (Using blobs for clarity)
X, _ = make_blobs(n_samples=50, centers=3, random_state=42)

# 2. CREATE THE LINKAGE MATRIX (The "Family Tree" logic)
# 'ward' linkage minimizes the variance of clusters being merged
Z = linkage(X, method='ward')

# 3. PLOT DENDROGRAM
plt.figure(figsize=(12, 5))
plt.title("Hierarchical Clustering Dendrogram")
dendrogram(Z)
plt.xlabel("Sample Index")
plt.ylabel("Distance (Ward)")
plt.axhline(y=10, color='r', linestyle='--') # Theoretical cut-off line
plt.show()

# 4. PERFORM CLUSTERING
# Based on the dendrogram, let's pick 3 clusters
cluster_model = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels = cluster_model.fit_predict(X)

# 5. QUALITY MEASURE
print(f"Silhouette Score for Hierarchical: {silhouette_score(X, labels):.4f}")
```
