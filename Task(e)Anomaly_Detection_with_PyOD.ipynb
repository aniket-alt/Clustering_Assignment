```python
# 1. Install PyOD
!pip install pyod -q

from pyod.models.iforest import IForest
from pyod.utils.data import generate_data
from pyod.utils.example import visualize

# 2. GENERATE DATA WITH OUTLIERS
# 200 training points, 10% are outliers
X_train, X_test, y_train, y_test = generate_data(n_train=200, n_test=100, n_features=2, contamination=0.1, random_state=42)

# 3. TRAIN ISOLATION FOREST
clf_name = 'IsolationForest'
clf = IForest()
clf.fit(X_train)

# 4. PREDICT
y_train_pred = clf.labels_  # binary labels (0: inlier, 1: outlier)
y_train_scores = clf.decision_scores_  # raw outlier scores

# 5. VISUALIZE
visualize(clf_name, X_train, y_train, X_test, y_test, y_train_pred,
          clf.predict(X_test), show_figure=True, save_figure=False)
```
